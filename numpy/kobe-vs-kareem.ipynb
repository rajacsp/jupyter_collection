```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la
```

```python
years = np.array([2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016])
games = [80,77,82,82,73,82,58,78,6,35,66]
points = np.array([35.4,31.6,28.3,26.8,27,25.3,27.9,27.3,13.8,22.3,17.6])
```

```python
def show_game_plot(years, games, points):

    fig = plt.figure(figsize=(12, 10))
    axs = fig.subplots(2, 1, sharex=True)

    # plot 1
    axs[0].plot(years, points, 'b.',  ms = 10)
    axs[0].set_title('Kobe Bryant, Points per Game')
    axs[0].set_ylim(0, 40)
    axs[0].grid(True)

    # plot 2
    axs[1].bar(years, games)
    axs[1].set_title('Kobe Bryant, Games Played')
    axs[1].set_ylim([0, 100])
    axs[1].grid(True)

    plt.show()
```

```python
show_game_plot(years, games, points)
```

```python
# dropping 2013-2014 as he was injured during that season

index_2014 = np.where(years == 2014)

print(index_2014)
```

```python
# dropping 2014

years = np.delete(years, index_2014)
games = np.delete(games, index_2014)
points = np.delete(points, index_2014)
```

```python
show_game_plot(years, games, points)
```
