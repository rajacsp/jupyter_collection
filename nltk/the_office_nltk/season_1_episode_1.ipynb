```python
def get_rap_custom_stopwords():
    list = ['na', '\'m', '\'s', 'n\'t', '\'re', '\'ve', '\'d', 'er']
    return list

def cleanup(s):
    stop = stopwords.words('english') + list(string.punctuation) + get_rap_custom_stopwords()
    sb = [i for i in word_tokenize(s.lower()) if i not in stop]
    return sb
```

```python
tokens = cleanup(text)
```

```python
tokens
```

```python
fdist = FreqDist(tokens)
```

```python
fdist.plot(25, cumulative=True)
```
