```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

```python
# Get Python Series from the Sheet

df_python = pd.read_csv('python_st_questions.csv')
ser_python = df_python['questions']

print(ser_python.shape)
```

```python
max_count = ser_python.count()+1 
x_values =  pd.Series(range(1, max_count))
```

```python
# Get Java Series from the Sheet

df_java = pd.read_csv('java_st_questions.csv')
ser_java = df_java['questions'][1: max_count]

print(ser_java.shape)
```

```python
# Use subplot

fig, ax = plt.subplots()
ax.plot(x_values, ser_python, 'k--', label='Python')
ax.plot(x_values, ser_java, 'x:', label='Java')

# Now add the legend with some customizations.
legend = ax.legend(loc='upper left')

plt.title('Java vs Python Questions Count since 2008')
```
